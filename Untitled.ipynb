```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
from time import time
import pandas as pd
import joblib

# Téléchargement des données EMNIST (lettres et chiffres)
print("Téléchargement des données EMNIST...")
# EMNIST balanced contient des chiffres et des lettres (majuscules et minuscules quand distinctes)
X, y = fetch_openml('emnist_balanced', version=1, return_X_y=True, parser='auto')

# Fonction pour mapper les chiffres/lettres aux classes EMNIST
def map_emnist_class_to_label(y):
    # Mapping selon le codage ASCII
    # Pour les classes EMNIST, les chiffres et lettres sont mappés selon l'ordre ASCII
    mapping = {}
    # Chiffres: 0-9 (classes 0-9)
    for i in range(10):
        mapping[i] = str(i)
    # Lettres majuscules: A-Z (classes 10-35)
    for i in range(26):
        mapping[i + 10] = chr(65 + i)  # 65 est le code ASCII pour 'A'
    # Lettres minuscules: a-z (quand distinctes des majuscules) (classes 36-61)
    for i in range(26):
        mapping[i + 36] = chr(97 + i)  # 97 est le code ASCII pour 'a'
    
    # Créer un mapping pour les classes actuelles (EMNIST balanced a 47 classes)
    actual_mapping = {}
    unique_classes = sorted(np.unique(y))
    for i, cls in enumerate(unique_classes):
        if i < len(mapping):
            actual_mapping[cls] = mapping[i]
    
    # Convertir les classes numériques en étiquettes
    return np.array([actual_mapping[cl] for cl in y])

# Prétraitement des données
X = X / 255.0  # Normaliser les valeurs entre 0 et 1
y = y.astype(int)  # Convertir les classes en entiers

# Convertir les classes numériques en étiquettes pour l'affichage
y_labels = map_emnist_class_to_label(y)

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test, y_train_labels, y_test_labels = train_test_split(
    X, y, y_labels, test_size=0.2, random_state=42)

print(f"Nombre d'exemples d'entraînement: {X_train.shape[0]}")
print(f"Nombre d'exemples de test: {X_test.shape[0]}")
print(f"Nombre de classes: {len(np.unique(y))}")

# Fonction pour afficher des exemples d'images
def show_examples(X, y_labels, n_examples=10):
    plt.figure(figsize=(20, 4))
    for i in range(n_examples):
        ax = plt.subplot(1, n_examples, i+1)
        # Rotation de 90 degrés et inversion horizontale (nécessaire pour EMNIST)
        img = X.iloc[i].values.reshape(28, 28).T
        plt.imshow(img, cmap='gray')
        plt.title(f"Label: {y_labels[i]}")
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.tight_layout()
    plt.show()

print("Visualisation de quelques exemples d'images EMNIST:")
show_examples(X_train, y_train_labels)

# Modèle MLP pour la classification
print("Entraînement du modèle...")
# Augmentation de la taille du réseau pour gérer plus de classes
mlp = MLPClassifier(
    hidden_layer_sizes=(200, 100),  # Augmentation de la complexité du réseau
    activation='relu',
    solver='adam',
    batch_size=128,
    learning_rate_init=0.001,
    max_iter=30,  # Augmentation du nombre d'époques
    verbose=True,
    random_state=42
)

start_time = time()
mlp.fit(X_train, y_train)
training_time = time() - start_time
print(f"L'entraînement a pris {training_time:.2f} secondes")

# Évaluation du modèle
y_pred = mlp.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Précision sur le jeu de test: {accuracy:.4f}")

# Conversion des prédictions numériques en étiquettes pour l'affichage
y_pred_labels = map_emnist_class_to_label(y_pred)

# Matrice de confusion des principales classes
plt.figure(figsize=(15, 15))
# Utiliser uniquement les 20 premières classes pour la lisibilité
num_classes_to_show = min(20, len(np.unique(y)))
cm = confusion_matrix(y_test[:1000], y_pred[:1000])[:num_classes_to_show, :num_classes_to_show]
unique_classes = sorted(np.unique(y))[:num_classes_to_show]
class_labels = [map_emnist_class_to_label([cl])[0] for cl in unique_classes]
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Prédiction')
plt.ylabel('Réalité')
plt.title('Matrice de confusion (20 premières classes)')
plt.show()

# Rapport de classification détaillé
print("Rapport de classification:")
print(classification_report(y_test, y_pred, target_names=map_emnist_class_to_label(np.unique(y))))

# Courbe de perte pendant l'entraînement
plt.figure(figsize=(12, 4))
plt.plot(mlp.loss_curve_)
plt.title('Courbe de perte pendant l\'entraînement')
plt.xlabel('Itérations')
plt.ylabel('Perte')
plt.grid(True)
plt.show()

# Fonction pour prédire et visualiser
def predict_and_visualize(X, y_true, y_true_labels, n_samples=20):
    indices = np.random.choice(len(X), n_samples, replace=False)
    X_samples = X.iloc[indices]
    y_true_samples = y_true.iloc[indices]
    y_true_label_samples = y_true_labels[indices]

    y_pred_samples = mlp.predict(X_samples)
    y_pred_label_samples = map_emnist_class_to_label(y_pred_samples)
    
    plt.figure(figsize=(20, 8))
    for i in range(n_samples):
        ax = plt.subplot(2, n_samples//2, i+1)
        # Rotation de 90 degrés et inversion horizontale (nécessaire pour EMNIST)
        img = X_samples.iloc[i].values.reshape(28, 28).T
        plt.imshow(img, cmap='gray')
        
        title_color = 'green' if y_pred_samples[i] == y_true_samples.iloc[i] else 'red'
        plt.title(f"P: {y_pred_label_samples[i]}\nR: {y_true_label_samples[i]}", color=title_color)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.tight_layout()
    plt.show()

print("Visualisation des prédictions sur des exemples aléatoires:")
predict_and_visualize(X_test, y_test, y_test_labels)

# Fonction pour analyser des erreurs spécifiques
def analyze_specific_errors(true_label, pred_label, X, y_true, y_true_labels):
    y_pred = mlp.predict(X)
    
    # Trouver les indices où les erreurs correspondant à nos critères se produisent
    errors = []
    for i in range(len(y_true)):
        if y_true.iloc[i] == true_label and y_pred[i] == pred_label:
            errors.append(i)

    if len(errors) == 0:
        print(f"Aucune erreur trouvée où {map_emnist_class_to_label([true_label])[0]} est prédit comme {map_emnist_class_to_label([pred_label])[0]}")
        return
    
    print(f"Trouvé {len(errors)} cas où {map_emnist_class_to_label([true_label])[0]} est prédit comme {map_emnist_class_to_label([pred_label])[0]}")
    
    n = min(10, len(errors))
    plt.figure(figsize=(20, 4))
    for i in range(n):
        idx = errors[i]
        ax = plt.subplot(1, n, i + 1)
        # Rotation de 90 degrés et inversion horizontale (nécessaire pour EMNIST)
        img = X.iloc[idx].values.reshape(28, 28).T
        plt.imshow(img, cmap='gray')
        plt.title(f"Réel: {y_true_labels.iloc[idx]}\nPrédit: {map_emnist_class_to_label([pred_label])[0]}")
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.tight_layout()
    plt.show()

print("Analyse des erreurs spécifiques:")
# Exemples d'analyse d'erreurs spécifiques (adaptez les chiffres selon vos besoins)
# Chercher des confusions entre lettres et chiffres similaires
# Par exemple: 0 et O, 1 et l, 5 et S, 8 et B
if 0 in np.unique(y) and 24 in np.unique(y):  # 0 et O
    analyze_specific_errors(0, 24, X_test, y_test, y_test_labels)
if 1 in np.unique(y) and 21 in np.unique(y):  # 1 et l
    analyze_specific_errors(1, 21, X_test, y_test, y_test_labels)

# Fonction pour tester une image spécifique
def test_specific_image(index, X, y_true, y_true_labels):
    # Récupérer l'image et sa vraie étiquette
    image = X.iloc[index].to_numpy()
    true_label = y_true.iloc[index]
    true_label_char = y_true_labels.iloc[index]
    
    # Faire la prédiction
    pred_proba = mlp.predict_proba([image])[0]
    pred_label = np.argmax(pred_proba)
    
    # Convertir la prédiction en caractère
    pred_label_char = map_emnist_class_to_label([pred_label])[0]
    
    # Afficher l'image et les probabilités
    plt.figure(figsize=(14, 6))
    
    # Image
    plt.subplot(1, 2, 1)
    # Rotation de 90 degrés et inversion horizontale (nécessaire pour EMNIST)
    img = image.reshape(28, 28).T
    plt.imshow(img, cmap='gray')
    title_color = 'green' if pred_label == true_label else 'red'
    plt.title(f"Prédit: {pred_label_char}, Réel: {true_label_char}", color=title_color)
    
    # Graphique des probabilités pour les 10 classes les plus probables
    plt.subplot(1, 2, 2)
    top_indices = np.argsort(pred_proba)[-10:][::-1]
    top_probs = pred_proba[top_indices]
    top_labels = [map_emnist_class_to_label([idx])[0] for idx in top_indices]
    
    plt.bar(range(len(top_labels)), top_probs)
    plt.xticks(range(len(top_labels)), top_labels)
    plt.xlabel('Caractère')
    plt.ylabel('Probabilité')
    plt.title('Top 10 des probabilités')
    
    plt.tight_layout()
    plt.show()

print("Test d'images spécifiques:")
test_specific_image(42, X_test, y_test, y_test_labels)
test_specific_image(120, X_test, y_test, y_test_labels)

# Fonction pour prédire un caractère à partir d'une image externe
def predict_character(image_path):
    from PIL import Image, ImageOps
    import cv2
    
    # Charger l'image
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print(f"Impossible de charger l'image: {image_path}")
        return
    
    # Prétraiter l'image
    # Redimensionner à 28x28
    img = cv2.resize(img, (28, 28))
    # Inverser les couleurs si nécessaire (texte noir sur fond blanc -> texte blanc sur fond noir)
    if np.mean(img) > 128:
        img = 255 - img
    
    # Normaliser
    img = img / 255.0
    
    # Rotation de 90 degrés dans le sens inverse des aiguilles d'une montre et inversion
    # horizontale pour correspondre au format EMNIST
    img = np.rot90(img, k=3)
    img = np.fliplr(img)
    
    # Aplatir l'image en vecteur
    img_vector = img.flatten()
    
    # Afficher l'image prétraitée
    plt.figure(figsize=(4, 4))
    plt.imshow(img, cmap='gray')
    plt.title("Image prétraitée")
    plt.axis('off')
    plt.show()
    
    # Faire la prédiction
    pred_proba = mlp.predict_proba([img_vector])[0]
    pred_label = np.argmax(pred_proba)
    pred_char = map_emnist_class_to_label([pred_label])[0]
    
    print(f"Caractère prédit: {pred_char} (classe {pred_label})")
    
    # Afficher les probabilités des 5 meilleures prédictions
    top_indices = np.argsort(pred_proba)[-5:][::-1]
    for idx in top_indices:
        char = map_emnist_class_to_label([idx])[0]
        prob = pred_proba[idx]
        print(f"{char}: {prob:.4f}")
    
    return pred_char

# Fonction pour lire plusieurs caractères d'une image avec segmentation
def read_text_from_image(image_path):
    import cv2
    
    # Charger l'image
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print(f"Impossible de charger l'image: {image_path}")
        return
    
    # Prétraitement de l'image
    # Binarisation
    _, binary = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    
    # Recherche de contours - cela permettra de localiser chaque caractère
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Tri des contours de gauche à droite
    bounding_boxes = [cv2.boundingRect(contour) for contour in contours]
    # Trier par position x (pour lire de gauche à droite)
    # Filtrer les contours trop petits
    min_area = 25  # Ajustez cette valeur selon vos besoins
    filtered_boxes = [(x, y, w, h) for x, y, w, h in bounding_boxes if w*h > min_area]
    sorted_boxes = sorted(filtered_boxes, key=lambda x: x[0])
    
    # Afficher l'image avec les contours
    plt.figure(figsize=(10, 6))
    img_with_contours = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    for x, y, w, h in sorted_boxes:
        cv2.rectangle(img_with_contours, (x, y), (x+w, y+h), (0, 255, 0), 2)
    plt.imshow(cv2.cvtColor(img_with_contours, cv2.COLOR_BGR2RGB))
    plt.title("Image avec caractères détectés")
    plt.axis('off')
    plt.show()
    
    # Reconnaître chaque caractère
    recognized_text = ""
    for i, (x, y, w, h) in enumerate(sorted_boxes):
        # Extraire le caractère
        char_img = binary[y:y+h, x:x+w]
        
        # Ajouter un peu de padding si nécessaire
        padding = 4
        char_img_padded = np.pad(char_img, padding, mode='constant', constant_values=0)
        
        # Redimensionner à 28x28
        char_img_resized = cv2.resize(char_img_padded, (28, 28))
        
        # Rotation de 90 degrés et inversion horizontale pour EMNIST
        char_img_rotated = np.rot90(char_img_resized, k=3)
        char_img_flipped = np.fliplr(char_img_rotated)
        
        # Normaliser
        char_img_norm = char_img_flipped / 255.0
        
        # Aplatir l'image
        char_vector = char_img_norm.flatten()
        
        # Prédire
        pred_label = mlp.predict([char_vector])[0]
        pred_char = map_emnist_class_to_label([pred_label])[0]
        
        recognized_text += pred_char
        
        # Afficher le caractère isolé et sa prédiction
        plt.figure(figsize=(2, 2))
        plt.imshow(char_img_norm, cmap='gray')
        plt.title(f"Prédit: {pred_char}")
        plt.axis('off')
        plt.show()
    
    print(f"Texte reconnu: {recognized_text}")
    return recognized_text

# Enregistrer le modèle
joblib.dump(mlp, 'emnist_mlp_model.joblib')
print("Modèle enregistré sous 'emnist_mlp_model.joblib'")

print("Exemple d'utilisation:")
print("Pour prédire un caractère à partir d'une image:")
print("predict_character('path/to/image.png')")
print("\nPour lire plusieurs caractères d'une image:")
print("read_text_from_image('path/to/text_image.png')")
```

```python
# Fonction pour lire plusieurs caractères d'une image avec segmentation
def read_text_from_image(image_path):
    import cv2
    
    # Charger l'image
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print(f"Impossible de charger l'image: {image_path}")
        return
    
    # Prétraitement de l'image
    # Binarisation
    _, binary = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    
    # Recherche de contours - cela permettra de localiser chaque caractère
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Tri des contours de gauche à droite
    bounding_boxes = [cv2.boundingRect(contour) for contour in contours]
    # Trier par position x (pour lire de gauche à droite)
    # Filtrer les contours trop petits
    min_area = 25  # Ajustez cette valeur selon vos besoins
    filtered_boxes = [(x, y, w, h) for x, y, w, h in bounding_boxes if w*h > min_area]
    sorted_boxes = sorted(filtered_boxes, key=lambda x: x[0])
    
    # Afficher l'image avec les contours
    plt.figure(figsize=(10, 6))
    img_with_contours = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    for x, y, w, h in sorted_boxes:
        cv2.rectangle(img_with_contours, (x, y), (x+w, y+h), (0, 255, 0), 2)
    plt.imshow(cv2.cvtColor(img_with_contours, cv2.COLOR_BGR2RGB))
    plt.title("Image avec caractères détectés")
    plt.axis('off')
    plt.show()
    
    # Reconnaître chaque caractère
    recognized_text = ""
    for i, (x, y, w, h) in enumerate(sorted_boxes):
        # Extraire le caractère
        char_img = binary[y:y+h, x:x+w]
        
        # Ajouter un peu de padding si nécessaire
        padding = 4
        char_img_padded = np.pad(char_img, padding, mode='constant', constant_values=0)
        
        # Redimensionner à 28x28
        char_img_resized = cv2.resize(char_img_padded, (28, 28))
        
        # Rotation de 90 degrés et inversion horizontale pour EMNIST
        char_img_rotated = np.rot90(char_img_resized, k=3)
        char_img_flipped = np.fliplr(char_img_rotated)
        
        # Normaliser
        char_img_norm = char_img_flipped / 255.0
        
        # Aplatir l'image
        char_vector = char_img_norm.flatten()
        
        # Prédire
        pred_label = mlp.predict([char_vector])[0]
        pred_char = map_emnist_class_to_label([pred_label])[0]
        
        recognized_text += pred_char
        
        # Afficher le caractère isolé et sa prédiction
        plt.figure(figsize=(2, 2))
        plt.imshow(char_img_norm, cmap='gray')
        plt.title(f"Prédit: {pred_char}")
        plt.axis('off')
        plt.show()
    
    print(f"Texte reconnu: {recognized_text}")
    return recognized_text

# Enregistrer le modèle
joblib.dump(mlp, 'emnist_mlp_model.joblib')
print("Modèle enregistré sous 'emnist_mlp_model.joblib'")

print("Exemple d'utilisation:")
print("Pour prédire un caractère à partir d'une image:")
print("predict_character('path/to/image.png')")
print("\nPour lire plusieurs caractères d'une image:")
print("read_text_from_image('path/to/text_image.png')")
```
